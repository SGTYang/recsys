```python
user_data['mem_birth_year'] = user_data['mem_birth_year'].astype('int')
user_data['mem_birth_year'] = user_data['mem_birth_year'].astype('str')
user_data['mem_birth_year']
```

```python
user_data['mem_birth_month'] = user_data['mem_birth_month'].astype('int')
user_data['mem_birth_month'] = user_data['mem_birth_month'].astype('str')
user_data['mem_birth_month']
```

```python
user_data['mem_birth_date'] = user_data[['mem_birth_year', 'mem_birth_month']].agg('-'.join, axis=1)
user_data['mem_birth_date']
```

```python
from datetime import datetime
def cal_age(birth_data):
    try:
        now = datetime.now()
        birthday = datetime.strptime(birth_data, "%Y-%m")
        difference = now - birthday
        age = difference.days // 365
        return int(age)
    except Exception as e:
        print(birth_data)
        print(e)
        return None


user_data['age'] = user_data['mem_birth_date'].apply(cal_age)
```

```python
user_data[user_data['age'].isna()]
```

```python
user_data = user_data[user_data['age'].notnull()]
```

```python
user_data['age'] = user_data['age'].astype('int')
```
